```python
import pandas as pd
apparel = pd.read_excel('apparel sales.xlsx')
camp = pd.read_excel('camp sales.xlsx')
competition = pd.read_excel('competition sales.xlsx')
```

```python
columns_to_drop = ['Customer Name', 'Customer Type', 'Customer Class']
apparel.drop(columns_to_drop, axis=1, inplace=True)
camp.drop(columns_to_drop, axis=1, inplace=True)
competition.drop(columns_to_drop, axis=1, inplace=True)
```

```python
apparel.rename(columns={2020: 'Apparel-2020', 2021: 'Apparel-2021', 2022: 'Apparel-2022'}, inplace=True)
camp.rename(columns={2020: 'Camp-2020', 2021: 'Camp-2021', 2022: 'Camp-2022'}, inplace=True)
competition.rename(columns={2020: 'Competition-2020', 2021: 'Competition-2021', 2022: 'Competition-2022'}, inplace=True)
```

```python
all_products = pd.merge(apparel, camp, on='Customer Number', how='outer')
all_products = pd.merge(all_products, competition, on='Customer Number', how='outer')
```

# C. Question 1 - Work Shown

- From 2020 to 2022, has Varsity Spirit been successful in moving customers into additional Product Segments within the Scholastic Customer Segment?
- As part of your summary, can you answer: What was the increase in the # of Customers who purchased in all three Product Segments from 2020 to 2022?

```python
all_products['2020 PS Count'] = all_products[['Apparel-2020', 'Camp-2020', 'Competition-2020']].apply(lambda x: ((x > 0) | x.notna()) & (x != 0.0)).astype(int).sum(axis=1)
all_products['2021 PS Count'] = all_products[['Apparel-2021', 'Camp-2021', 'Competition-2021']].apply(lambda x: ((x > 0) | x.notna()) & (x != 0.0)).astype(int).sum(axis=1)
all_products['2022 PS Count'] = all_products[['Apparel-2022', 'Camp-2022', 'Competition-2022']].apply(lambda x: ((x > 0) | x.notna()) & (x != 0.0)).astype(int).sum(axis=1)
```

```python
all_products['PS Growth'] = all_products['2022 PS Count'] - all_products['2020 PS Count']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

product_segment_growth_counts = pd.DataFrame(all_products['PS Growth'].value_counts()).sort_index()

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
custom_palette = sns.color_palette("husl", len(product_segment_growth_counts))
ax = sns.barplot(x=product_segment_growth_counts.index, y=product_segment_growth_counts['PS Growth'], palette=custom_palette)

# Set plot labels and title
plt.xlabel('Change in Product Segment Count from 2020 to 2022', fontsize=14)  # Replace with your x-axis label
plt.ylabel('Number of Customers', fontsize=14)  # Replace with your y-axis label
plt.title('Change in Product Segment Count with from 2020 to 2022', fontsize=16)

# Add labels to the top of the bars
for index, value in enumerate(product_segment_growth_counts['PS Growth']):
    ax.text(index, value + 1, str(value), ha='center', va='bottom', fontsize=10, color='black')
    
plt.xticks(fontsize=16)  # Adjust the font size for x-axis ticks
plt.yticks(fontsize=12)  # Adjust the font size for y-axis ticks
    
# Show the plot
plt.show()
```

```python
product_segment_growth_counts.index.name = 'Product Segment Change from 2020-2022'
product_segment_growth_counts = product_segment_growth_counts.rename(columns={'PS Growth': '# of Customers'})
total_customers = product_segment_growth_counts['# of Customers'].sum()
product_segment_growth_counts['% Per PS Count Category'] = (((product_segment_growth_counts['# of Customers']
                                                                          /total_customers) * 100).round(2).astype(str) + '%')
product_segment_growth_counts
```

```python
positive_PS = (product_segment_growth_counts['# of Customers'][1] + product_segment_growth_counts['# of Customers'][2] + product_segment_growth_counts['# of Customers'][3])/total_customers
negative_PS = (product_segment_growth_counts['# of Customers'][-3] + product_segment_growth_counts['# of Customers'][-2] + product_segment_growth_counts['# of Customers'][-1])/total_customers

(positive_PS * 100).round(2).astype(str) + '%', (negative_PS * 100).round(2).astype(str) + '%'
```

- Is it true that customers who are active in multiple Product Segments both spend more overall and spend more within each Product Segment?

```python
all_products['2020 Spending'] = (all_products['Apparel-2020'].fillna(0) +
                                all_products['Camp-2020'].fillna(0) +
                                all_products['Competition-2020'].fillna(0))

all_products['2021 Spending'] = (all_products['Apparel-2021'].fillna(0) +
                                all_products['Camp-2021'].fillna(0) +
                                all_products['Competition-2021'].fillna(0))

all_products['2022 Spending'] = (all_products['Apparel-2022'].fillna(0) +
                                all_products['Camp-2022'].fillna(0) +
                                all_products['Competition-2022'].fillna(0))

all_products['Total Spending'] = (all_products['2020 Spending'] +
                                  all_products['2021 Spending'] +
                                  all_products['2022 Spending'])

all_products['Total Apparel Spending'] = (all_products['Apparel-2020'].fillna(0) +
                                all_products['Apparel-2021'].fillna(0) +
                                all_products['Apparel-2022'].fillna(0))

all_products['Total Camp Spending'] = (all_products['Camp-2020'].fillna(0) +
                                all_products['Camp-2021'].fillna(0) +
                                all_products['Camp-2022'].fillna(0))

all_products['Total Competition Spending'] = (all_products['Competition-2020'].fillna(0) +
                                all_products['Competition-2021'].fillna(0) +
                                all_products['Competition-2022'].fillna(0))
```

```python
TwentyTwenty_PS_Count = all_products.groupby('2020 PS Count').mean()
TwentyTwenty_PS_Count['Total Spending']
```

```python
TwentyOne_PS_Count = all_products.groupby('2021 PS Count').mean()
TwentyOne_PS_Count['Total Spending']
```

```python
TwentyTwo_PS_Count = all_products.groupby('2022 PS Count').mean()
TwentyTwo_PS_Count['Total Spending']
```

```python
concatenated_df = pd.concat([TwentyTwenty_PS_Count, TwentyOne_PS_Count, TwentyTwo_PS_Count], keys=['2020', '2021', '2022'])
total_spending = concatenated_df['Total Spending']
total_spending = total_spending.reset_index()
total_spending.columns = ['Year', 'PS Count', 'Total Spending']

# Set the style for better aesthetics (optional)
sns.set(style="whitegrid")

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Total Spending', hue='PS Count', data=total_spending, palette=custom_palette)

# Set plot labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Spending', fontsize=14)
plt.title('Total Spending Across Years and Product Segment Counts', fontsize=16)

# Show the plot
plt.show()
```
